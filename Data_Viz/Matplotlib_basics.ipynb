```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#basic plotting
x = np.linspace(5,100,100) #5 to 100 and 100 points in between
y = np.linspace(5,1000,100) #5 to 1000 with 100 points in between

#functional way of doing things 
plt.plot(x,y)
plt.show()
```

```python
#sky is the limit
x = np.linspace(0,5,10)
y = np.linspace(3,6,10)

plt.plot(x,y,'r-',x,y**2,'b+',x,y**3,'g^')
plt.show()
```

```python
#we can use out own ticks on x and y axis rather than the default ones
x = np.linspace(-5,5,200)
y = np.sinc(x)
y_1 = np.cos(x)

plt.plot(x,y,color='r',label='sinc')
plt.plot(x,y_1,color='b',label='cosine')

plt.legend(loc='upper left',frameon=False)
#our custom x and y axis ticks
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,1])
plt.show()
```

## Object Oriented Way

```python
num_rows,num_cols = 5,5
fig,axes = plt.subplots(num_rows,num_cols,figsize=(10,8))
for i in range(num_rows):
    for j in range(num_cols):
        axes[i,j].plot(x,y*i+1)
        
plt.tight_layout()
plt.show()
```

```python
#load in the data
df = pd.read_csv('./global_sales_data/market_fact.csv')
df.head()
```

```python
#box plot is an univariate analysis in which we see how is the distribution of one particular variable
plt.boxplot(df['Order_Quantity'])
plt.show()
#the extreme lines represnet extreme values
#yellow line represents the median value
```

```python
#But if the distribution of a particular vairable is very Haphazard then
plt.boxplot(df['Sales'])
plt.show()
#lets try to analyse why this is happening
```

```python
df['Sales'].describe()
#not the diffrences between various quartile values
```

```python
#we need to scale the data before plotting
#we can use plt.yscale command

fig,axes = plt.subplots(1,2,figsize=(10,8))
axes[0].boxplot(df['Sales'])

axes[1].boxplot(df['Sales'])
plt.yscale('log')
plt.tight_layout()
plt.show()
```
