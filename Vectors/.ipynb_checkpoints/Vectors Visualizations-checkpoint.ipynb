## Vector  Intuition
Our whole life we have heard the same thing again and again that vectors are __physical quantities that have direction__.That is an alright enough definition but alas it never forces us to look inside the hood and understand what is really going on.This notebook is an attempt to change that methodology and see for real what geometrical significance does vectors and matrices hold or why is for example matrix addition defined the way it is defined.

Inspired by 3blue1brown

We will look at two definitions of vectors side by side.

1)Physics - quantities that have a sense of direction associated with them

2)Computer Sciences - an ordered list of numbers
So let us start by first drawing a vector on 2-D coordinate axis

```python
#import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#plot a vector 3i+4j
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1,color='r') #(startx,starty,endx,endy)
#set the limits on x and y
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
#for our smooth working lets define a function to plot vectors
def plot_vectors(vecs):
    colors = ['r','b','g','y']
    i = 0
    for vec in vecs:
        plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units='xy',angles='xy',scale=1,color=colors[i%len(colors)]) #(startx,starty,endx,endy)
        i+=1
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()
```

```python
#lets plot i and j vectors
i = np.asarray([0,0,1,0])
j = np.asarray([0,0,0,1])
plot_vectors([i,j])
```

## Vector Addition
Okay now lets visualize the triangle rule

```python
#we will define two vectors
vec1 = np.asarray([0,0,3,4]) #3i+4j
vec2 = np.asarray([0,0,4,-1]) #0i+8j
plot_vectors([vec1,vec2])
#now we as the theory says that in vector addition all the components of all vectors 
#along their respective axis will get add
#that is just element wise addition
vec3 = vec1+vec2 #submission
print('resultant vector is {}'.format(vec3))
#now this vec3 must satisfy the triangle law that if two vectors are represented as two sides of a triangle in magnitude and direction (taken in order)
#their resultant will be the third side of triangle taken in opp. order
#thus we need to move vec2 to end points of vec1 
vec2 = np.asarray([3,4,4,-1])
plot_vectors([vec1,vec2,vec3])
```

Vector subtraction can be performed on simmilar lines by writing a-b as a + (-b)

## Projection
Dot Product of two vectors is defined as 

$\vec{a}\cdot\vec{b} = |\vec{a}| |\vec{b}| \cos(\theta) = a_x b_x + a_y b_y$ 

Thus we can find projection of one vector on another by the formula

$a_b = |\vec{a}| \cos(\theta) = |\vec{a}|\frac{\vec{a}\cdot\vec{b}}{|\vec{a}||\vec{b}|} = \frac{\vec{a}\cdot\vec{b}}{|\vec{b}|}$

```python
#we will find projection of vec1 on vec2 
#but lets first start by plotting unit vector of vec2 
vec2 = np.asarray([0,0,4,-1]) #reset

mag_vec2 = np.sqrt(vec2[2]**2+vec2[3]**2)
unit_vec2 = vec2/mag_vec2
print(unit_vec2) #unit vector
print(np.sqrt(unit_vec2[2]**2+unit_vec2[3]**2)) #mag of unit vector
plot_vectors([vec2,unit_vec2])
vec1_2 = np.dot(vec1,unit_vec2) #lenght of projection
proj1_2 = unit_vec2*vec1_2 #unit vector scaled by length
plot_vectors([vec1,vec2,proj1_2])
```

## Basis , Span and other stuff
Now in  our 2-D coordinate system we were defining every vector as a combination of __i , j__ vectors but consider a coordinate system in which some vectors __u , w__ are considered as the basis vectors that is any vector defined in that vector space must be a linear combination of __u and w__ 

```python
u = vec1 #our basis vectors
w = vec2 #our basis vectors
#general vector r is defined as 
#r = a*u + b*w
#we are keeping a fixed as 2 and varying b and seeing locus of r
#and in second plt statement we are keeping b constant and varying a
cons = 2
for var in range(1,5):
    plt.quiver(0,0,cons*u,var*w,scale_units='xy',angles='xy',scale=1,color='r')
    plt.quiver(0,0,var*u,cons*w,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.show()
```

Thus we can see linear combination of __u and w__ can conver given varying values of a and b the whole vector space

__But is it always the case?__ or is it also dependent on __u and w?__

```python
##Suppose w is a linear combination of u itself
#for eq w = u/2
u = vec1
w = vec1/2
#now in linear combination w has lost its individuality and we are basically plotting diffrent linear combination of u
#which will squish down our range space to 1-D (vectors that can be achieved by linear combination of u and w)
for i in range(100):
    #diff values of a and b
    a = (np.random.random(1)-0.5)*3
    b = (np.random.random(1)-0.5)*3
    r = a*u + b*w
    plt.quiver(r[0],r[1],r[2],r[3],scale_units='xy',angles='xy',scale=1,color='g') #plot the resultant of linear combination
plt.quiver(0,0,u[2],u[3], scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,w[2],w[3], scale_units='xy', angles='xy', scale=1, color='b')

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```
